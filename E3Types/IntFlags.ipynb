IntFlags represent bitwise combinations of values backed by an integer, like an int-backed enum that can have multiple values at once. They can be passed to other functions that expect regular integers, but also provide the useful representations of an enum.

With the example of BomType1 and BomType2:

```python
from e3series.types import BomType1, BomType2

# Flag-value relation:
print(f"Value of {BomType1.EXP_BOM_DEVICE.name}: {BomType1.EXP_BOM_DEVICE.value:#0{10}x}")
print(f"Value of {BomType1.EXP_BOM_CONN.name}: {BomType1.EXP_BOM_CONN.value:#0{10}x}")
print(f"Value of {BomType1.EXP_BOM_TUBE.name}: {BomType1.EXP_BOM_TUBE.value:#0{10}x}")

# Combination:
combined_flag = BomType1.EXP_BOM_DEVICE | BomType1.EXP_BOM_CONN | BomType1.EXP_BOM_TUBE
print(f"Value of combined flag {combined_flag.name}: {combined_flag.value:#0{10}x}")

# Short Representation for debugging purposes:
print(f"Short representation of combined flag: {repr(combined_flag)}")

# Different bitwise operations are supported:
inverted_flag = ~combined_flag
print(f"Bitwise Inversion: {repr(inverted_flag)}\n(Hex: {inverted_flag.value:#0{10}x})")
print(f"Bitwise AND with EXP_BOM_DEVICE: {repr(combined_flag & BomType1.EXP_BOM_DEVICE)}")
print(f"Bitwise OR with EXP_BOM_ASSEMBLY: {repr(combined_flag | BomType1.EXP_BOM_ASSEMBLY)}")
print(f"Bitwise XOR with EXP_BOM_CONN: {repr(combined_flag ^ BomType1.EXP_BOM_CONN)}")
print("etc.")
```

IntFlags can be created directly from integers as returned from COM functions:

```python
random_int = 12345
flag_from_int = BomType1(random_int)
print(f"Flag created from integer {random_int}: {repr(flag_from_int)}")
```

## Example


This allows us to easily print the BomType returned from GetBomType in a human-readable way

```python
def format_bom_type(GetBomTypeReturn: tuple[int, tuple[int, ...]])-> str:
    return_code = GetBomTypeReturn[0]
    bom_type = (BomType1(GetBomTypeReturn[1][0]), BomType2(GetBomTypeReturn[1][1]))
    if return_code < 0:
        return f"Error: Return code {return_code}"
    msg = f"BomType1: {bom_type[0].name} ({bom_type[0].value:#0{10}x}), BomType2: {bom_type[1].name} ({bom_type[1].value:#0{10}x})"
    return msg
```

Now print the BomType of all devices in the project:

```python

import e3series

app = e3series.Application()
job = app.CreateJobObject()
dev = job.CreateDeviceObject()

(ret, all_devices) = job.GetAllDeviceIds()
if ret == 0:
    print(f"Error getting all device IDs: Return code {ret}")
    exit(1)

for dev_id in all_devices:
    dev.SetId(dev_id)
    device_name = dev.GetName()
    message = f"BomType of {device_name}: {format_bom_type(dev.GetBomType())}"
    print(message)
    app.PutMessage(message)
```
